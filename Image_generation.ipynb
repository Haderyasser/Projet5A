### Fonction de génération

Les masques, les backgrounds ainsi que les positions des chardons sont sélectionnés aléatoirement.

```python
def createxy(x,y):
    x_normal=4060
    y_normal=3050
    while (x_normal > 3800 or y_normal > 2900) or ((x_normal < 0 or y_normal < 0)) :
        x_normal=np.random.normal(x,200)
        y_normal=np.random.normal(y,200)
    return x_normal,y_normal
```

```python
def img_generation(mask_path,img_path,N):
    
    for i in range(0,N):     
        b_rand=randint(0,3)#randomiser l'image de fond
        
        nb_chardon=randint(5,8)#nombre de chardons
        nb_groupes=randint(3,5)# nombre de groupes de chardons
        background = Image.open(img_path+'background_0/background%d.jpg' %(b_rand))
        bg_w, bg_h = background.size
        
        file = open("generated_images/coordinates%d.txt" %(i),"a")
        file.truncate(0)
        for j in range(nb_groupes):
            x_rand=randint(0,bg_w)#randomiser la position
            y_rand=randint(0,bg_h)#randomiser la position
            for k in range(1,nb_chardon):
                m_rand=randint(0,30)#randomiser le mask
                angle_rand=randint(0,360)#randomiser l'angle de rotation
                foreground = Image.open(mask_path+'mask_0/mask%d.png' %(m_rand))
                foreground.rotate(angle_rand)
                masking_insa=Image.open(masking_insapath+'background%d.jpg' %(b_rand))
                x_normal=createxy(x_rand,y_rand)[0]
                y_normal=createxy(x_rand,y_rand)[1]
                while masking_insa.getpixel((int(x_normal),int(y_normal)))!=0 :
                    x_normal=createxy(x_rand,y_rand)[0]
                    y_normal=createxy(x_rand,y_rand)[1]
                m_w,m_h=foreground.size
                file.write(str(int(x_normal)) +" " +str(int(y_normal))+ " "+str(int(x_normal)+m_w) +" "
                +str(int(y_normal)+m_h)+"\n")

                background.paste(foreground, (int(x_normal), int(y_normal)),foreground)#retirer le "foreground" pour avoir le fond noir
        file.close()
        background.save('generated_images/generated%d.jpg' %(i))
```

## Création des XML

```python
import xml.etree.ElementTree as ET
import cv2
import numpy as np
import matplotlib.pyplot as plt

def get_annots_from_xml(path):
    """
    Récupère les annotations contenues dans un fichier .xml au format Pascal VOC.
    Retourne un array utilisable en entrée des modèles Tensorflow et des fonctions d'affichage.
    """
    tree = ET.parse(path)
    root = tree.getroot()
    
    size_x = int(root.findall('size')[0][0].text)
    size_y = int(root.findall('size')[0][1].text)
    objects_annot = root.findall('object')
    arr_box_annot = np.zeros((len(objects_annot), 4), dtype=np.float32)

    for i, box_annot in enumerate(root.findall('object')):
        arr_box_annot[i, :] = [int(box_annot[4][0].text) / size_y,
                               int(box_annot[4][1].text) / size_x,
                               int(box_annot[4][2].text) / size_y,
                               int(box_annot[4][3].text) / size_x]
    return arr_box_annot

def create_xml_from_detection(path_xml, detection_boxes, detection_scores,
                              seuil_detection, name_image, path_image, width, height, depth):
    """
    Génère un XML au format Pascal VOC contenant les annotations détectées avec une confiance supérieure au seuil.
    """
    # Création de la racine du document et des infos générales
    root = ET.Element("annotation")
    ET.SubElement(root, "folder").text = "patchs"
    ET.SubElement(root, "filename").text = str(name_image)
    ET.SubElement(root, "path").text = str(path_image)
    source = ET.SubElement(root, "source")
    ET.SubElement(source, "database").text = "Detection"
    size = ET.SubElement(root, "size")
    ET.SubElement(size, "width").text = str(width)
    ET.SubElement(size, "height").text = str(height)
    ET.SubElement(size, "depth").text = str(depth)
    ET.SubElement(root, "segmented").text = "0"
    
    # Nombre de bounding boxes détectées avec assez de confiance
    nb_boxes = np.sum(detection_scores >= seuil_detection)
    
    # Création de chaque bounding box
    # Le modèle renvoie les bounding boxes par ordre décroissant de confiance
    for i in range(nb_boxes):
        arr_boxes = np.array(detection_boxes)[i, :]
        obj = ET.SubElement(root, "object")
        ET.SubElement(obj, "name").text = "chardon"
        ET.SubElement(obj, "pose").text = "Unspecified"
        ET.SubElement(obj, "truncated").text = "0"
        ET.SubElement(obj, "difficult").text = "0"
        bndbox = ET.SubElement(obj, "bndbox")
        ET.SubElement(bndbox, "xmin").text = str(int(np.round(arr_boxes[0] * width)))
        ET.SubElement(bndbox, "ymin").text = str(int(np.round(arr_boxes[1] * height)))
        ET.SubElement(bndbox, "xmax").text = str(int(np.round(arr_boxes[2] * width)))
        ET.SubElement(bndbox, "ymax").text = str(int(np.round(arr_boxes[3] * height)))
                                                    
    # Assemblage de l'arbre
    tree = ET.ElementTree(root)
    
    # Ecriture du fichier
    tree.write(path_xml)
```

```python
nb_images=300 #mettre le nombre d'images dans le dossier generated_images
for i in range(nb_images):
    annots=np.loadtxt("generated_images/generated_same/coordinates%d.txt"%(i),"int")#mettre le dossier contenant les images
    create_xml_from_detection("generated_images/generated_same/generated%d.xml" %(i), annots, np.ones(annots.shape[0]),
                              0, "generated%d.jpg" %(i), "generated_images/generated_same/generated%d.jpg" %(i), 1, 1, 3)
```

```python
def GradientHor(x):
    y=x-np.roll(x,1,axis=1)
    y[:,0]=0
    return y
def GradientVer(x):
    y=x-np.roll(x,1,axis=0)
    y[0,:]=0
    return y
def DivHor(x):
    N=len(x[0])
    y=x-np.roll(x,-1,axis=1)
    y[:,0]=-x[:,1]
    y[:,N-1]=x[:,N-1]
    return y
def DivVer(x):
    N=len(x)
    y=x-np.roll(x,-1,axis=0)
    y[0,:]=-x[1,:]
    y[N-1,:]=x[N-1,:]
    return y
def Gradient(x):
    y=[]
    y.append(GradientHor(x))
    y.append(GradientVer(x))
    return y
def Div(y):
    x=DivHor(y[0])+DivVer(y[1])
    return x

def Proj(im,ma,iref):
    res = np.copy(iref) 
    mask = np.where(ma==1)
    res[mask]=im[mask]
    return res

def GradientFonc(x,y):
    res = Div(np.array(Gradient(x))-np.array(y))
    return res


def FISTAPoissonEditing(targ,sour,ma,step,alpha,Niter,y):
    x=np.copy(targ)
    mask=np.where(ma==1)
    x[mask]=sour[mask]
    f=[]
    x1 = x
    for i in range(0,Niter):
        x_tilde = x+i/(i+alpha)*(x-x1)
        x1 = x_tilde
        x = Proj(x_tilde - step * GradientFonc(x_tilde,y),ma,targ)
        f+=[npl.norm((np.array(Gradient(x))- y))**2]

    f=np.array(f)    
    return np.clip(x,0,255),f[10:]



def _mixing_gradient(target, source):

    grad_target = np.array(Gradient(target))
    norm_target = np.sqrt(grad_target[0]**2 + np.array(grad_target[1]**2)) # this gives a matrix = [norm(u,v)]_ij 

    grad_source = np.array(Gradient(source))
    norm_source = np.sqrt(grad_source[0]**2 + np.array(grad_source[1]**2)) # this gives a matrix = [norm(u,v)]_ij 
    
    targ_sup_sour = norm_target > norm_source 
    index = np.where(targ_sup_sour==True)
    #print(index[1].shape)
    #print(grad_source.shape)
    v = grad_source
    
    a = norm_target > norm_source
    for i, j in zip(range(grad_source.shape[0]),range(grad_source.shape[1])):
        if a[i,j]==True:
            v[i,j] = grad_target[i,j]
    return v
```
